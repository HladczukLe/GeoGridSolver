```python
import pandas as pd
```

```python
dataset = pd.read_csv("C:\\GeoGridSolver\\data\\world_data.csv", sep=";", encoding='latin1')
dataset.head()
```

```python
#In Europe and is Landlocked
geo = dataset.loc[(dataset['Region'] == "Europe") & (dataset['Coastal Distance']=="Landlocked")]
grid1 = geo[['Country','Region','Coastal Distance']]
grid1
```

```python
#In Europe and Population under 1 million
geo = dataset.loc[(dataset['Region']=="Europe") & (dataset['Population Country'] < 1000000)]
grid2 = geo[['Country','Region','Population Country']]
grid2
```

```python
#In Europe and Capital city population over 2 million
geo = dataset.loc[(dataset['Region'] == "Europe") & (dataset['Population Capital'] > 2000000)]
grid3 = geo[['Country','Region','Capital', 'Population Capital']]
grid3
```

```python
#Name is 10+ letters long and is Landlocked
geo = dataset.loc[(dataset['Country'].str.len() >= 10) & (dataset['Coastal Distance']=="Landlocked")]
grid4 = geo[['Country','Region','Coastal Distance']]
grid4
```

```python
#Name is 10+ letters long and Population under 1 million
geo = dataset.loc[(dataset['Country'].str.len() >= 10) & (dataset['Population Country'] < 1000000)]
grid5 = geo[['Country','Region','Population Country']]
grid5
```

```python
#Name is 10+ letters long and Capital city population over 2 million
geo = dataset.loc[(dataset['Country'].str.len() >= 10) & (dataset['Population Capital'] > 2000000)]
grid6 = geo[['Country','Region','Capital', 'Population Capital']]
grid6
```

```python
#Name consists of multiple words and is Landlocked
geo = dataset.loc[(dataset['Country'].str.contains(' ', regex=False)) & (dataset['Coastal Distance'] == "Landlocked")]
grid7 = geo[['Country','Region','Coastal Distance']]
grid7
```

```python
#Nome do pais com multiplas palavras e tem menos de 1 milhao de habitantes
geo = dataset.loc[(dataset['Country'].str.contains(' ', regex=False)) &  (dataset['Population Country'] < 1000000)]
grid8 = geo[['Country','Region','Population Country']]
grid8
```

```python
#Nome do pais com multiplas palavras e a capital tem mais de 2 milhoes de habitantes
geo = dataset.loc[(dataset['Country'].str.contains(' ', regex=False)) &  (dataset['Population Capital'] > 2000000)]
grid9 = geo[['Country','Region','Capital', 'Population Capital']]
grid9
```
