Some codes:

- Starts/Ends with letter 

``dataset.loc[(dataset['Country'].str.startswith('B'))]``

- Name is multiple words

``dataset.loc[(dataset['Country'].str.contains(' ', regex=False))]``

- Name is # letters long

``dataset.loc[(dataset['Country'].str.len() >= 10)]``



```python
import pandas as pd
```

```python
dataset = pd.read_csv("C:\\GeoGridSolver\\data\\world_data.csv", sep=";", encoding='latin1')
```

```python
#Countries that are duplicated have more than one capital
duplicated = dataset[dataset.duplicated(['Country'], keep=False)]
duplicated = duplicated[['Country','Capital']]
duplicated
```

```python
dataset.info()
```

```python
#Capital is not the most populated city and Population under 5 million
geo = dataset.loc[(dataset['Capital most pop'] == "No") & (dataset['Population Country']<5000000)]
grid1 = geo[['Country','Population Country']]
grid1
```

```python
#Capital is not the most populated city and Borders 5+ countries
geo = dataset.loc[(dataset['Capital most pop'] == "No") & (dataset['Number of borders']>=5)]
grid2 = geo[['Country','Region','Capital','Number of borders']]
grid2
```

```python
#Capital is not the most populated city and Is landlocked
geo = dataset.loc[(dataset['Capital most pop'] == "No") & (dataset['Coastal Distance (km)']=='Landlocked')]
grid3 = geo[['Country','Capital', 'Coastal Distance (km)']]
grid3
```

```python
#Name consists of multiple words and Population under 5 million
geo = dataset.loc[(dataset['Country'].str.contains(' ', regex=False) & (dataset['Population Country']< 5000000))]
grid4 = geo[['Country','Region','Population Country']]
grid4
```

```python
#Name consists of multiple words and Borders 5+ countries
geo = dataset.loc[(dataset['Country'].str.contains(' ', regex=False) & (dataset['Number of borders']>=5))]
grid5 = geo[['Country','Region','Number of borders']]
grid5
```

```python
#Name consists of multiple words and Is landlocked
geo = dataset.loc[(dataset['Country'].str.contains(' ', regex=False)) & (dataset['Coastal Distance (km)'] == 'Landlocked')]
grid6 = geo[['Country','Region','Coastal Distance (km)']]
grid6
```

```python
#Starts with letter B and Population under 5 million
geo = dataset.loc[(dataset['Country'].str.startswith('B')) & (dataset['Population Country'] < 5000000)]
grid7 = geo[['Country','Region', 'Population Country']]
grid7
```

```python
#Starts with letter B and Borders 5+ countries
geo = dataset.loc[(dataset['Country'].str.startswith('B')) & (dataset['Number of borders'] >=5)]
grid8 = geo[['Country','Region','Number of borders']]
grid8
```

```python
#Starts with letter B and Is Landlocked
geo = dataset.loc[(dataset['Country'].str.startswith('B')) &  (dataset['Coastal Distance (km)'] == "Landlocked")]
grid9 = geo[['Country','Region','Coastal Distance (km)']]
grid9
```
